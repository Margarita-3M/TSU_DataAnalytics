```python
import requests
import json
import os
import pandas as pd
```

```python
# print(os.path.abspath(os.getcwd()))
FileName = './dics/roles_hh.json'

def get_profession():
    """
    Создаем метод для получения страницы со списком профессий.
    """
     
    req = requests.get('https://api.hh.ru/professional_roles') # Посылаем запрос к API
    data = req.content.decode() # Декодируем его ответ, чтобы Кириллица отображалась корректно
    req.close()
    return data

js_obj = json.loads(get_profession())

with open(FileName, 'w', encoding='utf8') as f:
        f.write(json.dumps(js_obj, ensure_ascii=False))

print('Страницы поиска собраны')
```

```python
with open(FileName, "r", encoding='utf8') as read_file:
    data = json.load(read_file)

it_roles = data['categories'][7]['roles']
df_it_roles = pd.json_normalize(it_roles)
df_it_roles[['id', 'name']].to_csv('./dics/roles_it.csv', index=False) 
```
