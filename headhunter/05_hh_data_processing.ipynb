# Обработка сырых json'ов: выгрузка в файлы для дальнейшей работы

```python
import requests
import json
import os
import time
import random
import pandas as pd
import csv
import re
import numpy as np
```

```python
path_download = "D:/vacancies_descr"
```

```python
# создаем пустой датафрейм с общей информацией
df_general = pd.DataFrame(
    {
        'id': [],
        'name': [],
        'area_name': [], 
        'salary_from': [],
        'salary_to': [],
        'salary_ccy': [],
        'salary_gross': [],
        'type_name': [],
        'address_lat': [],
        'address_lng': [],
        'experience_name': [],
        'schedule_name': [],
        'employment_name': [],
        'employer_id': [], 
        'employer_name': [],
        'published_at': [],
        'alternate_url': [],        
    })

# создаем пустой датафрейм с описанием вакансий
df_descr = pd.DataFrame(
    {
        'id': [],
        'description': [],        
    })

# создаем списки для справочника key_skills
key_skills_vac = []
key_skills_name = []

# создаем списки для справочника specializations
spec_vac = []
spec_name = []

# создаем списки для справочника professional_roles
role_vac = []
role_name = []

count = 1

# Получаем перечень ранее созданных файлов со списком вакансий и проходимся по нему в цикле и заполняем датафрейм
for fl in os.listdir(path_download):
    with open(os.path.join(path_download, fl), 'r', encoding='utf8') as f_read:
        jsonText = f_read.read()
    
    v = json.loads(jsonText)
    
    print('{}й считан ({})'.format(count, fl))
    # заполняем df_general
    vacancy_details = {}
    
    if 'id' in v:
        vacancy_details['id'] = v['id']
        vacancy_details['name'] = v['name']
        vacancy_details['area_name'] = v['area']['name']
        vacancy_details['type_name'] = v['type']['name']
        vacancy_details['experience_name'] = v['experience']['name']
        vacancy_details['schedule_name'] = v['schedule']['name']
        vacancy_details['employment_name'] = v['employment']['name']
        vacancy_details['published_at'] = v['published_at']  
        vacancy_details['alternate_url'] = v['alternate_url'] 
    else:
        pass
    
    if ('salary' in v) and (v['salary'] is not None):
        if 'from' in v['salary']:
            vacancy_details['salary_from'] = v['salary']['from']
        else:
            vacancy_details['salary_from'] = np.nan

        if 'to' in v['salary']:
            vacancy_details['salary_to'] = v['salary']['to']
        else:
            vacancy_details['salary_to'] = np.nan

        if 'currency' in v['salary']:
            vacancy_details['salary_ccy'] = v['salary']['currency']
        else:
            vacancy_details['salary_ccy'] = np.nan  
            
        if 'gross' in v['salary']:
            vacancy_details['salary_gross'] = v['salary']['gross']
        else:
            vacancy_details['salary_gross'] = np.nan                  
    else:
        vacancy_details['salary_from'] = np.nan
        vacancy_details['salary_to'] = np.nan
        vacancy_details['salary_ccy'] = np.nan
        vacancy_details['salary_gross'] = np.nan

    if ('address' in v) and (v['address'] is not None):
        if 'lat' in v['address']:
            vacancy_details['address_lat'] = v['address']['lat']
        else:
            vacancy_details['address_lat'] = np.nan
            
        if 'lng' in v['address']:
            vacancy_details['address_lng'] = v['address']['lng']
        else:
            vacancy_details['address_lng'] = np.nan
    else:
        vacancy_details['address_lat'] = np.nan
        vacancy_details['address_lng'] = np.nan

    if ('employer' in v) and (v['employer'] is not None):
        if 'name' in v['employer']:
            vacancy_details['employer_name'] = v['employer']['name']
        else:
            vacancy_details['employer_name'] = np.nan
        
        if 'id' in v['employer']:
            vacancy_details['employer_id'] = v['employer']['id']
        else:
            vacancy_details['employer_id'] = np.nan   
    
    df_temp = pd.DataFrame(vacancy_details, index=[0])
    df_general = df_general.append(df_temp, ignore_index=True)
    # =================================
    
    # заполняем df_descr
    vacancy_details = {}
    vacancy_details['id'] = v['id']
    vacancy_details['description'] = v['description']
    df_temp = pd.DataFrame(vacancy_details, index=[0])
    df_descr = df_descr.append(df_temp, ignore_index=True)
    # =================================
    
    # заполняем key_skills
    for skill in v['key_skills']:
        key_skills_vac.append(v['id'])
        key_skills_name.append(skill['name'])
    # =================================
    
    # заполняем specializations
    for spec in v['specializations']:
        spec_vac.append(v['id'])
        spec_name.append(spec['name'])
    # =================================
    
    # заполняем professional_roles
    for role in v['professional_roles']:
        role_vac.append(v['id'])
        role_name.append(role['name'])               
    # =================================
    
    print('{}й обработан ({})'.format(count, fl))
    count += 1
    
# заполняем датафрейм key_skills 
df_key_skills = pd.DataFrame({'id': key_skills_vac, 'key_skill': key_skills_name})        
# заполняем датафрейм specializations       
df_spec = pd.DataFrame({'id': spec_vac, 'specialization': spec_name}) 
# заполняем датафрейм professional_roles       
df_role = pd.DataFrame({'id': role_vac, 'prof_role': role_name}) 

print('Вакансии обработаны')
```

```python
path_save = 'D:/vac_info'
```

```python
df_general.to_csv(os.path.join(path_save, 'vac_general.csv'))
df_descr.to_csv(os.path.join(path_save, 'vac_descr.csv'))
df_key_skills.to_csv(os.path.join(path_save, 'vac_key_skills.csv'))
df_role.to_csv(os.path.join(path_save, 'vac_role.csv'))
df_spec.to_csv(os.path.join(path_save, 'vac_spec.csv'))
```
